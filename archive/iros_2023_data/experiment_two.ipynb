Show monotonic relationship between proposed psuedo ground truth metrics and groun truth.

```python
%cd ../..
from run_scripts.run import run_sweep

import os
import subprocess
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import matplotlib_inline
kpse_cp = subprocess.run(['kpsewhich', '-var-value', 'TEXMFDIST'], capture_output=True, check=True)
font_loc1 = os.path.join(kpse_cp.stdout.decode('utf8').strip(), 'fonts', 'opentype', 'public', 'tex-gyre')
print(f'loading TeX Gyre fonts from "{font_loc1}"')
font_dirs = [font_loc1]
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

plt.rcParams['font.family'] = 'TeX Gyre Termes'
plt.rcParams["mathtext.fontset"] = "stix"

matplotlib_inline.backend_inline.set_matplotlib_formats("svg")

large_dataset_name_mac = "r1-five-straight-3rounds.json"
large_dataset_name_wh4 = "p1_WH4_2.json"
small_dataset_name_mac = "r1-five-straight-elevator-3rounds.json"
small_dataset_name_wh4 = "4_lou_to_rich_room.json"
pso = 1
```

```python
osr = run_sweep(large_dataset_name_mac, pso)
gt_results = osr.gt_results_list
alpha_results = osr.alpha_results_list
shift_results = osr.shift_metric_list

plt.plot(gt_results, alpha_results, 'o')
plt.xlabel("Ground Truth Metric")
plt.ylabel("Alpha Metric")
plt.show()
plt.plot(gt_results, shift_results, 'o')
plt.xlabel("Ground Truth Metric")
plt.ylabel("Shift Metric")
plt.show()
```
