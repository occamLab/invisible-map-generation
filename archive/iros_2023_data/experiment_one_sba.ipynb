# Experiment One (SBA)
The purpose of this experiment is to evaluate the relative accuracy and success rate of the tuned hyperparameters for a given map. This evaluation will be done in three main methods:
1. Determine the lowest ground truth value for a given map based on a parameter sweep.
2. Determine if the tuned hyperparameters of one map can be extrapolated to that of another map that either contains this map or is contained within this map.
3. Determine if there is a smooth relationship between a swept hyperparameter and the ground truth data.

Each of these specific tests will be conducted twice - once with SBA optimization and once without.
